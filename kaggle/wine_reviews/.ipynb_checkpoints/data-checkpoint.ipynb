```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size': 16})
import math

import seaborn as sns

import os
```

```python
abs_path = os.path.join('~', 'Development', 'kaggle_data', 'wine_reviews')

filename_data1 = os.path.join(abs_path, 'winemag-data_first150k.csv')
filename_data2 = os.path.join(abs_path, 'winemag-data-130k-v2.csv')

df1 = pd.read_csv(filename_data1)
df2 = pd.read_csv(filename_data2)
```

```python
df = pd.concat([df1, df2], sort=False)
df.head()
df = df.dropna(subset=['price'])
```

```python
group = df.groupby('country')['price'].count()
sorted_group = group.sort_values(axis=0, ascending=False)

values = sorted_group.values
print(sorted_group.index)
print(len(sorted_group.index))

other_countries = [sorted_group.index[10:]]
```

```python
# Make new column
# Very important and difficult grammar
df['new_country'] = np.where(np.isin(df['country'] ,other_countries), 'Others', df['country'])
```

```python
group_count = df.groupby('new_country').count().sort_values('country', ascending=False)
print(group_count)
order_countries = group_count.index
print(order_countries)
```

```python
country_mapping = {c: i for i, c in enumerate(order_countries) }
df['index_country'] = df['new_country'].map(country_mapping)
```

```python
g = sns.countplot(x='new_country', data=df, order=order_countries)
loc, labels = plt.xticks()
g.set_xticklabels(labels, rotation=90)
g.set(xlabel='Country', ylabel='Number of Reviews')
```

```python
g = sns.barplot(x='new_country', y='price', data=df, order=order_countries)
loc, labels = plt.xticks()
g.set_xticklabels(labels, rotation=90)
g.set(xlabel='Country', ylabel='Price')
```

```python
g = sns.barplot(x='new_country', y='points', data=df, order=order_countries)
loc, labels = plt.xticks()
g.set_xticklabels(labels, rotation=90)
g.set(xlabel='Country', ylabel='Points')
```

```python
df['cprice'] = df['price']
ps = np.arange(0, 101, 20)
for i, p in enumerate(ps):
    df.loc[(df['cprice']>p) & (df['cprice']<=p+20), 'cprice'] = int(i)
df.loc[df['cprice'] >= 100, 'cprice'] = 10
df['cprice'] = df['cprice'].astype(int)
```

```python
group = df.groupby('cprice').mean()
std = df.groupby('cprice').std()
plt.errorbar(x=group['price'], y=group['points'], xerr=std['price'], yerr=std['points'], fmt='bo')
plt.xlim(left=0, right=110)
plt.ylim(bottom=82, top=95.)
plt.xlabel('Price')
plt.ylabel('Points')
```
